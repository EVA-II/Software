```python
class DataProcessor:
    def __init__(self, raw_data):
        """
        Args:
            raw_data: 原始数据集（需是三维数组格式）
            seq_length: 序列长度（根据 `create_inout_sequences` 中的逻辑）
        """
        self.raw_data = raw_data
        self.train = None
        self.valid = None
        self.test = None

    def split_dataset(self, train_scale=0.8, valid_scale=0.1):
        """划分数据集"""
        data = self.raw_data.to_numpy()
        total = len(data)
        train_num = int(train_scale * total)
        valid_num = int(valid_scale * total)
        
        self.train = data[:train_num,:]
        self.valid = data[train_num:train_num+valid_num,:]
        self.test = data[train_num+valid_num:,:]
        return self.train, self.valid, self.test
    
    def create_inout_sequences(self,input_data, tw):
        self.input_seq = []
        L = int(len(input_data)/tw)
        j=0
        #print(L)
        for i in range(L):
            train_seq = input_data[j:j+tw]
            #train_label = input_data[i+tw:i+tw+1]
            #inout_seq.append((train_seq ,train_label))
            j=j+2742
            self.input_seq.append(train_seq )
        return self.input_seq
```

```python
class mydata(Dataset):
    def __init__(self,Data_input,Data_output):
        self.Data_input = torch.tensor(Data_input, dtype=torch.float32)
        self.Data_output = torch.tensor(Data_output, dtype=torch.float32)
    def __len__(self):
        return len(self.Data_input)
    def __getitem__(self, index):
        return self.Data_input[index],self.Data_output[index]
```

```python
class LSTM(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, output_size, batch_size):
        super().__init__()
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        self.output_size = output_size
        self.num_directions = 1 # 单向LSTM
        self.batch_size = batch_size
        # 注意：这里 `batch_first=True`，因此输入的形状是 (batch_size, seq_len, input_size)
        self.lstm = nn.LSTM(self.input_size, self.hidden_size, self.num_layers, batch_first=True)
        self.linear = nn.Linear(self.hidden_size, self.output_size)

    def forward(self, input_seq):
        # input_seq 应该是 (batch_size, seq_len, input_size)
        # 这里使用 .shape 获取维度
        #print(input_seq.shape[0])
        #print(input_seq.shape[1])
        batch_size = input_seq.shape[0]  # 从输入数据中获取 batch_size
        seq_len = input_seq.shape[1]      # 从输入数据中获取 seq_len
        #batch_size, seq_len = input_seq[0], input_seq[1]
        h_0 = torch.zeros(self.num_directions * self.num_layers, input_seq.shape[0], self.hidden_size).to(device)
        c_0 = torch.zeros(self.num_directions * self.num_layers, input_seq.shape[0], self.hidden_size).to(device)
        #print(h_0.shape)
        # output(batch_size, seq_len, num_directions * hidden_size)
        output, _ = self.lstm(input_seq, (h_0, c_0))
        print(output.shape)
        pred = self.linear(output)
        
        #
        #pred = pred[:, -1, :]
        #print(pred.shape)
        #pred = pred.reshape(-1,output_size)
        
        return pred
```

```python
args = argparse.Namespace(
    input_size=1,
    hidden_size=30,
    num_layers=2,
    output_size=1,
    bidirectional=False,
    batch_size=8,
    optimizer='adam',
    lr=0.005,
    weight_decay=5e-4,
    step_size=200,
    gamma=0.9,
    epochs=1500
)
```

```python
train_loader = torch.utils.data.DataLoader(tr_dataset,
                                               batch_size=args.batch_size, shuffle=False,
                                               num_workers=0,pin_memory=True)
validate_loader = torch.utils.data.DataLoader(va_dataset,
                                                  batch_size=1, shuffle=False,
                                                  num_workers=0,pin_memory=True)
test_loader = torch.utils.data.DataLoader(te_dataset,
                                                  batch_size=1, shuffle=False,
                                                  num_workers=0,pin_memory=True)
```

```python
input_size, hidden_size, num_layers = args.input_size, args.hidden_size, args.num_layers
output_size = args.output_size

if args.bidirectional:
    model = BiLSTM(input_size, hidden_size, num_layers, output_size, batch_size=args.batch_size).to(device)
else:
    model = LSTM(input_size, hidden_size, num_layers, output_size, batch_size=args.batch_size).to(device)

loss_function = nn.SmoothL1Loss(reduction='mean').to(device)
if args.optimizer == 'adam':
    optimizer = torch.optim.Adam(model.parameters(), lr=args.lr)
else:
    optimizer = torch.optim.SGD(model.parameters(), lr=args.lr, momentum=0.9)

scheduler = StepLR(optimizer, step_size=args.step_size, gamma=args.gamma)

# training
min_epochs = 10
start_epoch = 0
running_epoch_loss = []
val_avg_loss = []
best_model = None

# ==========================================
# ### 早停机制：参数初始化 ###
# ==========================================
patience = 80  # 容忍度：连续 20 个 epoch 验证集 loss 不降则停（可按需修改或设为 args.patience）
early_stop_counter = 0  # 计数器
min_val_loss = float('inf')  # 建议设为正无穷，确保第一个有效 loss 能被记录

total_step = 0
alpha = 0

for epoch in range(start_epoch, args.epochs):
    model.train()
    running_loss = 0.0
    for step, data in enumerate(train_loader):
        total_step = total_step + 1
        seq, label = data
        seq = seq.to(device)
        label = label.to(device)
        
        optimizer.zero_grad()
        y_pred = model(seq)
        
        loss_data = loss_function(y_pred, label)
        loss_dis = loss_data
        loss_dis.backward()
        optimizer.step()
        running_loss += loss_dis.item() * seq.shape[0]
        
    running_epoch_loss1 = running_loss / train_num
    running_epoch_loss.append(running_epoch_loss1)
    scheduler.step()
    
    # validation
    model.eval()
    total_loss = 0.0
    with torch.no_grad():
        for step, val_data in enumerate(validate_loader):
            val_input, val_output = val_data
            val_output_predict = model(val_input.to(device))
            val_loss_dis = loss_function(val_output_predict, val_output.to(device))
            total_loss = total_loss + val_loss_dis.item() * val_input.shape[0]
            
    val_avg_loss1 = total_loss / val_num
    
    # ==========================================
    # ### 早停机制：核心逻辑 ###
    # ==========================================
    if epoch > min_epochs:
        if val_avg_loss1 < min_val_loss:
            min_val_loss = val_avg_loss1
            best_model = copy.deepcopy(model)
            early_stop_counter = 0  # 发现更优的loss，计数器清零
        else:
            early_stop_counter += 1  # loss没有变优，计数器+1
            
    val_avg_loss.append(val_avg_loss1)
    
    # 绘图和打印部分保持不变
    if (epoch + 1) % 10 == 0:
        clear_output(wait=True)  # 清理旧的输出
        print('=====================================================')
        print(f"Epoch [{epoch}/{args.epochs}], 训练集loss: {running_epoch_loss[-1]:.6f}, 测试集loss: {val_avg_loss[-1]:.6f}")
        
        plt.plot(range(epoch + 1), running_epoch_loss, 'r-', lw=2, label='训练集Loss') 
        plt.plot(range(epoch + 1), val_avg_loss, 'g-', lw=2, label='测试集Loss') 
        plt.legend()
        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.title('Loss Over Epochs')
        plt.grid(True)
        plt.pause(0.1) 

    # ==========================================
    # ### 早停机制：触发终止 ###
    # ==========================================
    if early_stop_counter >= patience:
        print(f'\n>>> 触发早停机制！在第 {epoch} 个 Epoch 停止训练。连续 {patience} 个 Epoch 验证集损失未下降。')
        break

# 保存模型
if best_model is not None:
    state = {'models': best_model.state_dict()}
    torch.save(state, 'best_model_r.pth')
    print('已保存最佳模型权重。')
else:
    print('警告：未保存任何模型（可能是因为设定的 min_epochs 过大或验证集损失一直没达标）。')
```
